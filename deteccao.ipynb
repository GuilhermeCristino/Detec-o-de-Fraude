```python
#Importando bibliotecas
import pandas as pd
import numpy as np 
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,mean_squared_error,r2_score
```

```python
#Importando arquivo 
df = pd.read_csv('creditcard.csv')
```

```python
#Verificando Dataframe
df.head
```

```python
#Verificando o Tipo dos Dados
df.dtypes
```

```python
#Consultando Valores nulos
df.isnull().sum()
```

```python
#Verificando tipos dos dados NÂO FRAUDE
df_nao_fraude = df.Amount[df.Class == 0]

df_nao_fraude.describe()
```

```python
#Verificando os Tipos dos Dados de FRAUDE
df_fraude = df.Amount[df.Class == 1]

df_fraude.describe()
```

```python
#Contagem FRAUDE/NÂO FRAUDE
df.Class.value_counts()
```

```python
#FRAUDE
df_fraude = df[df.Class == 1]
df_fraude
```

```python
#NÃO FRAUDE
df_nao_fraude = df[df.Class == 0]
df_nao_fraude

```

```python
#Igualando a quantidade de FRAUDE e NÂO FRAUDE
df_nao_fraude = df_nao_fraude.sample(n=492)
df_nao_fraude
```

```python
#Contatenando FRAUDE/NÃO FRAUDE
df = pd.concat([df_nao_fraude, df_fraude], axis=0)
df
```

```python
#Ajustando Index
df.reset_index(inplace=True)
df
```

```python
#Retirando Dados para Validação
df_val_nao_fraude = df.head(5)
df_val_nao_fraude
```

```python
#Retirando Dados para Validação
df_val_fraude = df.tail(5)
df_val_fraude
```

```python
#Retirando as 5 primeiras Linhas da Validação
df = df.iloc[5:]

#Retirando as 5 ultimas Linhas da Validação
df = df[:-5]

df
```

```python
#Concatenando dataframes de validação
df_val_total = pd.concat([df_val_nao_fraude,df_val_fraude])

df_val_total.reset_index(inplace=True)
df_val_total_real = df_val_total.Class
df_val_total = df_val_total.drop(['level_0','index','Time','Class'], axis=1)
df_val_total
```

```python
# Verificando a distribuição dos dados 
df.Class.value_counts()
```

```python
# Separar Labels e Features
X = df.drop(['index','Time','Class'], axis=1)
Y = df['Class']
```

```python
print(X,Y)
```

```python
#Separando dados de Treino e Teste 
X_train, X_teste, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42, stratify=Y)
```

```python
#Treinamento
lr = LogisticRegression(max_iter=1000)
lr.fit(X_train, Y_train)
pred = lr.predict(X_teste)
acc = accuracy_score(Y_test, pred)

f'Acurácia:{acc * 100: .2f}'
```

```python
#Validação
pred = lr.predict(df_val_total)

df = pd.DataFrame({'real' :df_val_total_real, 'previsao':pred})

print(df)
```
